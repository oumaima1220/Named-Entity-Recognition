```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2)
```

```python
from collections import Counter
print(Counter(y_train))
print(Counter(y_test))
```

#building up train data and test data

```python
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["Word"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["Word"],"labels":y_test})
```

```python
train_data
```

```python
test_data
```

Model Training

```python
from simpletransformers.ner import NERModel,NERArgs
```

```python
label = data["labels"].unique().tolist()
label
```

```python
args = NERArgs()
args.num_train_epochs = 29
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32
```

Creating the NERModel, with  model_type (bert)  and a model_name (bert-base-cased)





```python
model = NERModel('bert', 'bert-base-cased',labels=label,args =args,use_cuda=True)
```

```python
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```python
result, model_outputs, preds_list = model.eval_model(test_data)
```

```python
result
```
